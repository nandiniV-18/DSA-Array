```python
df.rename(columns={'idea': 'description'}, inplace=True)
df.head()
```

```python
data=df[["deal","description"]]
display(data)
```

```python
df_secured = data[data['deal'] != 'No Deal']
df_notsecured = data[data['deal'] == 'No Deal']
```

```python
df_secured.head()
```

```python
df_notsecured.head()
```

```python
df_secured = df_secured.drop(columns=['deal'])
df_notsecured = df_notsecured.drop(columns=['deal'])
```

```python
df_notsecured.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
all_stopwords=stopwords.words('english')
sw_list = ['also', 'made', 'makes', 'like', 'this', 'even', 'company']
all_stopwords.extend(sw_list)

print(all_stopwords)
```

```python
df_secured["chars"]=df_secured["description"].str.len()
df_notsecured["chars"]=df_notsecured["description"].str.len()
sum_secured=df_secured["chars"].sum()
sum_notsecured=df_notsecured["chars"].sum()
print("True Corpus", sum_secured)
print("False Corpus", sum_notsecured)
```

```python
display(df_secured)
```

```python
display(df_notsecured)
```

```python
import pandas as pd
import numpy as np
column_array = df_secured['description'].values
split_words = [word.split() for word in column_array]
flattened_words = [word for sublist in split_words for word in sublist]
nsw_secured = np.array(flattened_words)
print(nsw_secured)
```

```python
import pandas as pd
import numpy as np
column_array = df_notsecured['description'].values

split_words = [word.split() for word in column_array]
flattened_words = [word for sublist in split_words for word in sublist]

nsw_notsecured = np.array(flattened_words)

print(nsw_notsecured)
```

```python
freq=nltk.FreqDist(nsw_secured)
freq_false=freq.most_common(10)
print(freq_false)
```

```python
freq=nltk.FreqDist(nsw_notsecured)
freq_false=freq.most_common(10)
print(freq_false)

```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# # # Download the stopwords from NLTK
# import nltk
# nltk.download('stopwords')
# nltk.download('punkt')
# Remove words from nsw_secured that are in all_stopwords
nsw_secured = np.array([word for word in nsw_secured if word not in all_stopwords])

# Display the filtered array
print(nsw_secured)
```

```python
freq=nltk.FreqDist(nsw_secured)
freq_false=freq.most_common(10)
print(freq_false)

```

```python
freq=nltk.FreqDist(nsw_notsecured)
freq_false=freq.most_common(10)
print(freq_false)
```
